```python
from keras.applications import VGG16, VGG19, Xception, ResNet50, InceptionV3
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from keras.preprocessing.image import ImageDataGenerator
from keras import models
from keras import layers
from keras import optimizers
```

```python
data128 = np.load('img_features128.npy')
```

```python
plt.imshow(data128[0])
```

## VGG16
256

```python
# weightsは重みのチェックポイント。include_topは全結合層を含めるかどうか。imagenetの1000クラス分類に対応。
conv_base = VGG16(weights = "imagenet",
                 include_top=False,
                 input_shape=(128,128,3))

hidden = conv_base.predict(data128)

input1 = layers.Input((4, 4, 512,))
flat1 = layers.Flatten(name='flat1')(input1)
dens1 = layers.Dense(256, name='hidden')(flat1)

model = models.Model(inputs=input1, outputs=dens1)
model.compile(loss='categorical_crossentropy',
              optimizer='Adam',
              metrics=['accuracy'])

features_vgg16 = model.predict(hidden)

features_vgg16.shape
```

```python
img_vgg16_256_df = pd.DataFrame(features_vgg16)
img_vgg16_256_df.to_csv('img_vgg16_256_df.csv')
```
